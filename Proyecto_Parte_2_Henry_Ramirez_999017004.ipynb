*Unificacion de la data y remplazo de los valores null*

```python
#Unificacion de los 16 dataframe
data_completa = pd.concat([
    data_2024, data_2023, data_2022, data_2021, data_2020,
    data_2019, data_2018, data_2017, data_2016, data_2015,
    data_2014, data_2013, data_2012, data_2011, data_2010,
    data_2009
], ignore_index=True)


#Eliminacion de la columna num_correlativo
data_completa = data_completa.drop(columns=["num_correlativo"], errors="ignore")
```

*Remplazo de los valores null*

```python
# Reemplazo de valores por defecto "IGNORADO" según el diccionario de datos:
#   Solo se aplican condiciones a las columnas g_hora, g_hora_5 y g_modelo_veh,
#   ya que sus valores dependen de la información de otros campos.
data_completa["g_hora"] = (
    data_completa["hora_ocu"].apply(
        lambda x: 1 if 0 <= x <= 5 else
                  2 if 6 <= x <= 11 else
                  3 if 12 <= x <= 17 else
                  4 if 18 <= x <= 24 else 5
    )
)



data_completa["g_hora_5"] = data_completa["g_hora"].apply(
    lambda x: 1 if x in [1,2] else 2 if x==3 else 3 if x==4 else 4
)



data_completa["g_modelo_veh"] = data_completa["modelo_veh"].apply(
    lambda x: 1 if 1970 <= x <= 1979 else
              2 if 1980 <= x <= 1989 else
              3 if 1990 <= x <= 1999 else
              4 if 2000 <= x <= 2009 else
              5 if 2010 <= x <= 2019 else
              6 if 2020 <= x <= 2029 else 99
)



data_completa["grupo_color"] = data_completa["color_veh"].apply(
    lambda x: 1 if x in [3, 4, 5, 9, 10, 15] else       # 1--> Colores Oscuros
              2 if x in [2, 8, 11, 12, 13, 16] else     # 2--> Colores Claros
              3 if x in [1, 6, 7, 14] else              # 3--> Colores Vivos
              4 if x == 17 else                         # 4--> Otros colores
              5                                         # 5 ---> 'Desconocidos'
)

```

# **Redes Neuronales 1: Clasificación de hecho de tránsito (colisión y no colisión) para motocicletas**

```python
#Filtramos la data con los parametro establecido
data_red_n1 = data_completa[(data_completa["tipo_veh"] == 4) & (data_completa["depto_ocu"] <= 22)]
```

```python
#Se crea una columna si el evento es colision es 1 si no es entonces es 0
data_red_n1.loc[:, 'colision'] = data_red_n1['tipo_eve'].apply(lambda x: 1 if x == 1 else 0)
```

```python
data_red_n1 = data_red_n1.sample(frac=1, random_state=42).reset_index(drop=True)
```

**Definición de variables independientes y dependientes (a predecir); definción de set de entreno y testeo**

```python
x_rn1 = data_red_n1[["color_veh","hora_ocu","g_modelo_veh"]]
x_rn1 = x_rn1.fillna(0)
y_rn1 = data_red_n1['colision']
```

```python
X_train_rn1, X_test_rn1, y_train_rn1, y_test_rn1 = train_test_split(x_rn1, y_rn1, test_size=0.2, random_state=42)
```

**Diseño y entreno de la red neuronal**

```python
model_rn1 = Sequential()
model_rn1.add(Dense(8, input_dim=3, activation='relu'))
model_rn1.add(Dense(4, activation='relu'))
model_rn1.add(Dense(1, activation='sigmoid'))
```

```python
model_rn1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model_rn1.fit(X_train_rn1, y_train_rn1, epochs=100, batch_size=64, validation_data=(X_test_rn1, y_test_rn1))
```

```python
loss_rn1, accuracy_rn1 = model_rn1.evaluate(X_test_rn1, y_test_rn1)
print(f'Loss: {loss_rn1}, Accuracy: {accuracy_rn1}')
```

**Predicción**

```python
hecho_rn1 = np.array([[1,     # 1 --> Color blanco
                  9,      # Hora de ocurrencia
                  6 ]])  # 6--> Modelo 2020 - 2029


res_rn1 = model_rn1.predict(hecho_rn1)
print(res_rn1)
```
