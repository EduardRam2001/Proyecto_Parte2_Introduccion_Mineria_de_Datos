*Unificacion de la data y remplazo de los valores null*

```python
#Unificacion de los 16 dataframe
data_completa = pd.concat([
    data_2024, data_2023, data_2022, data_2021, data_2020,
    data_2019, data_2018, data_2017, data_2016, data_2015,
    data_2014, data_2013, data_2012, data_2011, data_2010,
    data_2009
], ignore_index=True)


#Eliminacion de la columna num_correlativo
data_completa = data_completa.drop(columns=["num_correlativo"], errors="ignore")
```

*Remplazo de los valores null*

```python
# Reemplazo de valores por defecto "IGNORADO" según el diccionario de datos:
#   Solo se aplican condiciones a las columnas g_hora, g_hora_5 y g_modelo_veh,
#   ya que sus valores dependen de la información de otros campos.
data_completa["g_hora"] = (
    data_completa["hora_ocu"].apply(
        lambda x: 1 if 0 <= x <= 5 else
                  2 if 6 <= x <= 11 else
                  3 if 12 <= x <= 17 else
                  4 if 18 <= x <= 24 else 5
    )
)



data_completa["g_hora_5"] = data_completa["g_hora"].apply(
    lambda x: 1 if x in [1,2] else 2 if x==3 else 3 if x==4 else 4
)



data_completa["g_modelo_veh"] = data_completa["modelo_veh"].apply(
    lambda x: 1 if 1970 <= x <= 1979 else
              2 if 1980 <= x <= 1989 else
              3 if 1990 <= x <= 1999 else
              4 if 2000 <= x <= 2009 else
              5 if 2010 <= x <= 2019 else
              6 if 2020 <= x <= 2029 else 99
)



data_completa["grupo_color"] = data_completa["color_veh"].apply(
    lambda x: 1 if x in [3, 4, 5, 9, 10, 15] else       # Grupo 1: colores 3,4,5,9,10,15
              2 if x in [2, 8, 11, 12, 13, 16] else     # Grupo 2: colores 2,8,11,12,13,16
              3 if x in [1, 6, 7, 14] else              # Grupo 3: colores 1,6,7,14
              4 if x == 17 else                         # Grupo 4: color 17
              5                                         # Grupo 5: cualquier otro valor
)

```
